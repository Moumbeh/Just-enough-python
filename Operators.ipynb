```python
# Arithmetic operator
x = 3
y = 6

total = x + y
print(total)

total = x - y
print(total)

total = x * y
print(total)

total = y/x
print(total)

total = y % x
print(total)

total = y**x
print(total)
```

```python
# comparism operators
a = 10
b = 15
answer = a < b
print(answer)

answer = (a > b)
print(answer)

answer = (a == b)
print(answer)

answer = (a <= b)
print(answer)

answer = (a >= b)
print(answer)

answer = a != b
print(answer)
```

```python
# Assignment operators
c = 1
d = 2

c+=d # c = c+d
print(c)

c-=d # c = c-d
print(c)
```

```python
# logical operator
# and
# or

a = 20
b = 40

x = 7
y = 9

out = (a < b) or (x > y)
print(out)

out = (a > b) or (x < y)
print(out)

out = (a > b) or (x > y) # if you are true on either side then is true
print(out)

out = (a > b) and (x > y)
print(out)

out = (a < b) and (x < y) # And operator both sides has to be true for it to be true
print(out)

out = not(x < y) # this does the opposite
print(out)
```

```python
# Membership operators
first_tuple = ("DevOps", "Azure","Bash", 47, 60, 1.5)
ans = "DevOps" in first_tuple
print(ans)

ans = "DevOps" not in first_tuple
print(ans)

ans = 14 not in first_tuple
print(ans)
```
